```python
import pandas as pd
import numpy as np
import copy
df = pd.read_csv('./preprocessed_data.csv',low_memory=False)
```

```python
fr = df.groupby(['episode_id', 'event_type'])['event_type'].count().unstack().reset_index().fillna(0).set_index('episode_id')
```

```python
fr[fr > 1] = 1
```

```python
fr
```

```python
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
from mlxtend.frequent_patterns import fpgrowth
```

```python
frequent_itemsets_fr = apriori(fr, min_support=0.02, use_colnames=True)
frequent_itemsets_fr.sort_values(by='support',ascending=False).head(10)
```

```python
# Create the rules
rules_fr = association_rules(frequent_itemsets_fr, metric="lift", min_threshold=1)
rules_fr.head()
```

```python
fpgrowth(fr, min_support=0.1, use_colnames=True)
```

                MINIMUM SUPPORT

A float between 0 and 1 for minimum support of the itemsets returned. The support is computed as the fraction transactions_where_item(s)_occur / total_transactions.
